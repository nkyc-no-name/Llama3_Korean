```python
import json

# JSON 파일 읽기
with open('민원(콜센터) 질의응답_금융보험_상품 가입 및 해지_Training.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
all = []
dialog = []

for turn in data:
  n = int(turn['문장번호'])-1
  s = turn['화자']
  if n == 0:
    all.append(dialog); dialog = []

  cq = turn["고객질문(요청)"]
  rq = turn["상담사질문(요청)"]
  ca = turn["고객답변"]
  ra = turn["상담사답변"]

  txt = cq+rq+ca+ra

  if s == '고객':
    dialog.append({"from": "human", "value": txt})
  else:
    dialog.append({"from": "gpt", "value": txt})

all = all[1:]
```

```python
from datasets import Dataset
dataset = Dataset.from_dict({"conversations": all})
dataset
```
