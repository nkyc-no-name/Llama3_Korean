```python
# !pip install -q huggingface_hub
import huggingface_hub
import json
from datasets import Dataset, load_dataset
import os
import regex
```

```python
def json_to_hub (file_path, hub_repo, subset, split):

    # JSON 파일 읽기
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)

    all = []
    dialog = []

    for turn in data:
      n = int(turn['문장번호'])-1
      s = turn['화자']
      if n == 0:
        all.append(dialog); dialog = []

      cq = turn["고객질문(요청)"]
      rq = turn["상담사질문(요청)"]
      ca = turn["고객답변"]
      ra = turn["상담사답변"]

      txt = cq+rq+ca+ra

      if s == '고객':
        dialog.append({"from": "human", "value": txt})
      else:
        dialog.append({"from": "gpt", "value": txt})

    all = all[1:]
    dataset = Dataset.from_dict({"conversations": all})

    huggingface_hub.login("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")
    dataset.push_to_hub(hub_repo, subset, split=split)
```

```python
file_path_train = './aihub_minwonQA/training/'
file_path_valid = './aihub_minwonQA/validation/'

fn_train=os.listdir(file_path_train)
fn_valid=os.listdir(file_path_valid)

for file in fn_train:
  idx = regex.search('_Training.json', file)
  n = idx.start()
  json_to_hub(file_path_train+file, "NoNameFactory/callcenter", file[13:n], split="train")

for file in fn_valid:
  idx = regex.search('_Validation.json', file)
  n = idx.start()
  json_to_hub(file_path_valid+file, "NoNameFactory/callcenter", file[13:n], split="valid")
```
